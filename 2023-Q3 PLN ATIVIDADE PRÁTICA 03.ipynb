#### **CONFIGURANDO A AUTENTICAÇÃO COM A API**

Para acessar a API do Spotify, é necessário antenticar o aplicativo com as credenciais apropriadas. Logo a baixo estará o passo a passo de como configurar a autenticação

1) Obtenção das Credenciais do Spotify
    1. Acesse o Spotify Developer Dashboard: https://developer.spotify.com/
    2. Faça login com sua conta do Spotify (crie uma se necessário).

2) Criando um aplicativo
    1. No dashboard do Desenvolvedor do Spotify, clique em: 'Crie um Aplicativo'
    2. Forneça um nome e uma descrição do aplicativo

3. Obtenção das credenciais de autenticação
    1. Após criar o aplicativo, você verá as credenciais necessárias:  
        - Client ID
        - Client Secret
    2. Anote essas informações em um local seguro, elas serão utilizadas para autenticar seu aplicativo com a API.

4) Autenticando-se no notebook do Google Colab
    - No seu notebook, utilize o código seguinte:

```
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Substitua 'client_id' e 'client_secret' pelas suas credenciais do Spotify
client_id = 'client_id'
client_secret = 'client_secret'

#Configura a autenticação com as credenciais
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
```

Com essas estapas você terá configurado a autenticação do seu aplicativo com a API do Spotify no Google Colab.

#### **UTILIZAÇÃO BÁSICA COM EXEMPLO PRÁTICO**

1) Fazendo uma solicitação simples
    - Para fazer uma solicitação à API do Spotify, você pode usar o objeto **'sp'** (que configuramos anteriormente). 
    - Vamos fazer uma busca de informações sobre um álbum específico, utilizaremos o ID do álbum:

```
#Substitua 'album_id' pelo ID do álbum que você deseja buscar
album_id = 'album_id_aqui'

#Solicitação para obter informações sobre o álbum
album_info = sp.album(album_id)
```

2) Processando a resposta da API
    - A resposta será um dicionário JSON contendo as informações sobre o álbum, utilize o código:

```
# Extraindo informações da resposta
nome_album = album_info['name']
artista = album_info['artists'][0]['name']
data_lançamento = album_info['release_date']
total_de_faixas = album_info['total_tracks']

# Imprimindo as informações
print(f'Nome do Álbum: {nome_album}')
print(f'Artista: {artista}')
print(f'Data de Lançamento: {data_lançamento}')
print(f'Total de Faixas: {total_de_faixas}')
```
