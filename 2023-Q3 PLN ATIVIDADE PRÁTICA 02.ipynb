>

Um `programa Python` que utilize `expressões regulares` pode ajudar a identificar vários **tipos de erros** comuns em **livros**, especialmente erros de formatação e problemas relacionados à consistência do texto. Aqui estão alguns exemplos de erros comuns que podem ser identificados usando expressões regulares:

* Erros de gramática e ortografia: erros de digitação, concordância verbal e nominal, uso incorreto de pontuação e outros erros gramaticais.

* Problemas de formatação: você pode usar expressões regulares para encontrar erros de formatação, como espaços em excesso, tabulações inadequadas ou alinhamentos inconsistentes.

* Abreviações e acrônimos: você pode usar expressões regulares para encontrar abreviações ou acrônimos que não foram definidos ou explicados anteriormente no texto.

* Citações e referências: expressões regulares podem ser úteis para localizar citações ou referências que precisam de formatação especial.

* OUTROS TIPOS DE ERROS: não considerem apenas os tipos de erros citados acima.


**IMPORTANTE:** Lembre-se de que expressões regulares podem ser poderosas, mas também complexas. Dependendo da complexidade dos erros que você deseja identificar, pode ser necessário ajustar as expressões regulares de acordo com as características específicas do seu texto. Além disso, é importante ter em mente que as expressões regulares podem não ser a melhor ferramenta para todos os tipos de erros em livros, especialmente problemas mais contextuais ou semânticos, que podem exigir abordagens de PLN mais avançadas.

```
#Fiz essa função para verificar as 10 palavras (com mais de 5 letras) que mais aparece no capítulo 11
import requests
from bs4 import BeautifulSoup
import re
from collections import Counter

# Faz a requisição da página
url = 'https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap11/cap11.html'

# Faz a solicitação HTTP para obter o conteúdo HTML da página do livro
response = requests.get(url)

# Cria um objeto BeautifulSoup para parsear o HTML
soup = BeautifulSoup(response.content, 'html.parser')

# Encontra o elemento HTML que contém o texto do livro
book_text = soup.find('body')

# Extrai o texto do livro a partir do elemento HTML
book = book_text.get_text()

# Dividi o texto em palavras
words = re.findall(r'\b\w+\b', book)

# Filtra palavras com mais de 5 letras
palavras = [word for word in words if len(word) > 5]

# Conta a frequência das palavras
frequencia = Counter(palavras)

# Encontra as 10 palavras mais repetidas
top_10_words = frequencia.most_common(10)

# Imprimi o top 10 das palavras mais repetidas
for i, count in top_10_words:
    print(f'{i}: {count} vezes')
```

```
#Verifica se o espaçamento entre as palavras estão corretos. (se há algum espaçamento duplo, triplo, etc)
import requests
from bs4 import BeautifulSoup
import re

# URLs a serem verificadas
urls = ['https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap11/cap11.html',
        'https://brasileiraspln.com/livro-pln/1a-edicao/parte7/cap14/cap14.html']

for url in urls:
  # Faz a solicitação HTTP para obter o conteúdo HTML da página do livro
    response = requests.get(url)

    # Cria um objeto BeautifulSoup para parsear o HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontra o elemento HTML que contém o texto do livro
    book_text = soup.find('body')

    # Extrai o texto do livro a partir do elemento HTML
    book = book_text.get_text()

    # Dividir o texto em palavras
    words = re.findall(r'\b\w+\b', book)

    # Verifica se há espaços em branco repetidos entre as palavras
    erro = False
    for i in range(len(words) - 1):
        if '  ' in words[i]:
            print(f"Erro de formatação na URL '{url}': Existem espaços em branco repetidos entre as palavras '{words[i]}' e '{words[i + 1]}'\n")
            erro = True

    if not erro:
        print(f"Não há erro de formatação em espaçamento na URL '{url}'\n")
```

```
#Verifica a ortografia da palavra 'relação'
import requests
from bs4 import BeautifulSoup
import re

# Lista de URLs para verificar
urls = [
    'https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap11/cap11.html',
    'https://brasileiraspln.com/livro-pln/1a-edicao/parte7/cap14/cap14.html'
]

# Palavras corretas
palavras_corretas = ['relação', 'Relação']

# Expressão regular para encontrar variações de "relação"
expressao = r'[Rr]ela[cç][aãáà][õoóô]'

for url in urls:
    # Faz a requisição da página
    response = requests.get(url)

    # Cria um objeto BeautifulSoup para parsear o HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontra o elemento HTML que contém o texto do livro
    book_text = soup.find('body')

    # Extrai o texto do livro a partir do elemento HTML
    book = book_text.get_text()

    # Encontra todas as correspondências no texto
    correspondencias = re.findall(expressao, book, flags=re.IGNORECASE)

    # Verifica se há correspondências incorretas
    correspondencias_incorretas = [match for match in correspondencias if match not in palavras_corretas]

    print(f'URL: {url}')
    if correspondencias_incorretas:
        print("Variações incorretas da palavra 'relação' encontradas no texto:")
        for match in correspondencias_incorretas:
            print(match)
            print()
    else:
        print("Não foi encontrada nenhuma variação incorreta da palavra 'relação'.\n")   
```

```
#Verifica se as datas estão no padrão "certo": dd/mm/aaaa ou dd-mm-aaaa
import requests
from bs4 import BeautifulSoup
import re

# Lista de URLs para verificar
urls = [
    'https://brasileiraspln.com/livro-pln/1a-edicao/parte6/cap11/cap11.html',
    'https://brasileiraspln.com/livro-pln/1a-edicao/parte7/cap14/cap14.html'
]

# Expressão regular para encontrar datas no formato DD/MM/AAAA ou DD-MM-AAAA
expressao = r'^(?!\d{1,2}[/-]\d{1,2}[/-]\d{4}).*$'

# Variável para controlar se foram encontrados erros
erros_encontrados = False

for url in urls:
    # Faz a requisição da página
    response = requests.get(url)

    # Cria um objeto BeautifulSoup para parsear o HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontra o elemento HTML que contém o texto do livro
    book_text = soup.find('body')

    # Extrai o texto do livro a partir do elemento HTML
    book = book_text.get_text()

    # Usa expressões regulares para encontrar datas no texto
    datas = re.findall(expressao, book)

    # Verifica se foram encontradas datas
    if datas:
        print(f"Erro de formatação: Foram encontradas datas nas seguintes URLs: {url}")
        erros_encontrados = True

# Verifique se não foram encontrados erros
if not erros_encontrados:
    print("Nenhum erro de data foi encontrado nos capítulos 11 e 14.")
```
